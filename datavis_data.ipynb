Polar plot is used for the visualization of circular/directional component. Representation of data points as its distance from the centre and their angle or direction from a reference point.

Here , Euthanasia total has been visualised on a radial axis .(The 111 notation is a shorthand way of specifying this configuration. It is equivalent to explicitly calling add_subplot(1, 1, 1))
